```
# Get necessary library
%matplotlib inline
import matplotlib.pyplot as plt

# prepare data
x = [1,2,3,4]
y = [11,22,33,44]

#setup plot
fig,ax = plt.subplots(figsize=(10,10))

# plot data
ax.plot(x,y);

# customize plot
ax.set(title = "simple plot",
       xlabel = "x-axis",
       ylabel = "y_axis")

# save and show (we have to save the whole figure)
fig.savefig('/content/Sample_plot.png')
```

##TWO OPTION FOR SUBPLOT
###1) subplot 1

```
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,
                                          ncols=2,
                                          figsize=(10,5))
ax1.plot(x,x);
ax2.scatter(np.random.random(10),np.random.random(10));
ax3.bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax4.hist(x);
```

###2) subplot 2

```
fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(10,5))
#plots to each different index
ax[0,0].plot(x,x);
ax[0,1].scatter(np.random.random(10),np.random.random(10));
ax[1,0].bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax[1,1].hist(x);
```

#Which one should you use ?(pyplot vs matplotlib OO method)
##* when plotting quickly use pyplot method
##* when advanced plotting use OO method

```
ov_50 = data2[data2["age"]>50]
ov_50
```

##Pyplot method

```
ov_50.plot(kind='scatter',
          x='age',
          y='chol',
          c='target');
```

## OO method with pyplot method

```
fix, ax = plt.subplots(figsize=(10,6))
ov_50.plot(kind='scatter',
          x='age',
          y='chol',
          c='target',
           ax=ax);
#ax.set_xlim([45,100])
```

#OO (Object Oreintated) method from scratch

```
fig,ax = plt.subplots(figsize=(10,6))
#plot
scatter = ax.scatter(x=ov_50['age'],
                     y=ov_50['chol'],
                     c=ov_50["target"]);
#Customize
ax.set(title="Heart Disease & Cholestorol",
       xlabel="Age",
       ylabel="Cholestorol");
#Legend
ax.legend(*scatter.legend_elements(),title="Target");
#add a horizontal line
# ax.axhline(ov_50["chol"].mean(),ls= '--')
# OR
ax.axhline(ov_50["chol"].mean(),linestyle= '--');
```

```
ov_50.head()
```

### 2 - y-plots v/s 1-x-plot

```
fig, (ax1,ax2) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10,10))
#Add data to ax1
scatter = ax1.scatter(x=ov_50["age"],
                      y=ov_50["chol"],
                      c=ov_50["target"])
#Customize
ax1.set(title='age v/s cholestorol',
        xlabel='age',
        ylabel='chol')
#Legend
ax1.legend(*scatter.legend_elements(),title="Target");

#meanline
ax1.axhline(y=ov_50["chol"].mean(),
            ls="--")

#Add data to ax2
scatter = ax2.scatter(x=ov_50["age"],
                      y=ov_50["thalach"],
                      c=ov_50["target"])
#Customize
ax2.set(title='age v/s thalach',
        xlabel='age',
        ylabel='thalach')
#Legend
ax2.legend(*scatter.legend_elements(),title="Target");

#meanline
ax2.axhline(y=ov_50["thalach"].mean(),
            ls="--");
#add a figure title
fig.suptitle("Heart Disease Analysis",fontsize=16, fontweight="bold");
```

##Share x axis

```
fig, (ax1,ax2) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10,10),
                              sharex=True)   #with this statement
#Add data to ax1
scatter = ax1.scatter(x=ov_50["age"],
                      y=ov_50["chol"],
                      c=ov_50["target"])
#Customize
ax1.set(title='age v/s cholestorol',
        # xlabel='age',                      #we can ignore statement
        ylabel='chol')
#Legend
ax1.legend(*scatter.legend_elements(),title="Target");

#meanline
ax1.axhline(y=ov_50["chol"].mean(),
            ls="--")

#Add data to ax2
scatter = ax2.scatter(x=ov_50["age"],
                      y=ov_50["thalach"],
                      c=ov_50["target"])
#Customize
ax2.set(title='age v/s thalach',
        xlabel='age',
        ylabel='thalach')
#Legend
ax2.legend(*scatter.legend_elements(),title="Target");

#meanline
ax2.axhline(y=ov_50["thalach"].mean(),
            ls="--");
#add a figure title
fig.suptitle("Heart Disease Analysis",fontsize=16, fontweight="bold");
```

```
#Change style
# plt.style.use("seaborn-whitegrid")
plt.style.use("seaborn-whitegrid")

fig, (ax1,ax2) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10,10),
                              sharex=True)   #with this statement
#Add data to ax1
scatter = ax1.scatter(x=ov_50["age"],
                      y=ov_50["chol"],
                      c=ov_50["target"],
                      cmap = 'winter')
#Customize
ax1.set(title='age v/s cholestorol',
        # xlabel='age',                      #we can ignore statement
        ylabel='chol')
#Legend
ax1.legend(*scatter.legend_elements(),title="Target");

#meanline
ax1.axhline(y=ov_50["chol"].mean(),
            ls="--")

#Add data to ax2
scatter = ax2.scatter(x=ov_50["age"],
                      y=ov_50["thalach"],
                      c=ov_50["target"],
                      cmap='autumn')
#Customize
ax2.set(title='age v/s thalach',
        xlabel='age',
        ylabel='thalach')
#Legend
ax2.legend(*scatter.legend_elements(),title="Target");

#meanline
ax2.axhline(y=ov_50["thalach"].mean(),
            ls="--");
#add a figure title
fig.suptitle("Heart Disease Analysis",fontsize=16, fontweight="bold");
```

```
#Customizing the x and y axis limitations
fig, (ax1,ax2) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10,10),
                              sharex=True)   #with this statement
#Add data to ax1
scatter = ax1.scatter(x=ov_50["age"],
                      y=ov_50["chol"],
                      c=ov_50["target"],
                      cmap='winter')
#Customize
ax1.set(title='age v/s cholestorol',
        # xlabel='age',                      #we can ignore statement
        ylabel='chol')
# change limits for x axis
ax1.set_xlim([50,80])
#Legend
ax1.legend(*scatter.legend_elements(),title="Target");

#meanline
ax1.axhline(y=ov_50["chol"].mean(),
            ls="--")

#Add data to ax2
scatter = ax2.scatter(x=ov_50["age"],
                      y=ov_50["thalach"],
                      c=ov_50["target"],
                      cmap='winter')
#Customize
ax2.set(title='age v/s thalach',
        xlabel='age',
        ylabel='thalach')
#Legend
ax2.legend(*scatter.legend_elements(),title="Target");
# change ylim
ax2.set_ylim([60,200])
#meanline
ax2.axhline(y=ov_50["thalach"].mean(),
            ls="--");
#add a figure title
fig.suptitle("Heart Disease Analysis",fontsize=16, fontweight="bold");
```
